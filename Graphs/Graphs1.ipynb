```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10,0.1)
a = np.cos(x)
b = np.sin(x)
```

```python
plt.plot(x,a,'b', ls = 'dashed', marker = 'd', markersize = 5,markeredgecolor = 'k')   # (x-axis, y-axis, color)
plt.plot(x,b,'r',ls = 'dashdot', marker = 'o',markersize = 5, markeredgecolor = 'k')
plt.xlabel('Values of x')
plt.ylabel('Values of function(x)')
plt.show()
```

```python
plt.scatter(x,a,marker='x',s=3,c='r')
```

```python
plt.scatter(x,b)
```

```python
sub = ['Maths','Computer Science','English','Hindi']
marks = [90,70,50,80]
marks1 = [80,45,65,78]
marks2 = [10,20,30,40]
s=np.arange(4)
plt.bar(s,marks, color =['r','k','b','m'],width=0.25)
plt.bar(s+0.25,marks1, color =['m','b','k','r'],width=0.25)
plt.bar(s+0.50,marks2, color =['y','g','m','c'],width=0.25)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.xticks(s,sub)
```

```python
sub = ['Maths','Computer Science','English','Hindi']
marks = [90,70,50,80]
plt.barh(sub,marks)
plt.ylabel("Subjects")
plt.xlabel("Marks")
plt.title("Chart")
plt.show()
```

```python
sub = ['Maths','Computer Science','English','Hindi']
marks = [90,70,50,80]
plt.pie(marks, labels = sub, colors = ['red', 'cyan', 'pink', 'yellow'], explode=[0.1,0,0.2,0.3], autopct = "%05.3f%%")
plt.show()
```
